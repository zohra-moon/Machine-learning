```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
df=pd.read_csv("task_1_data.csv")
df.head()
```

```python
x = df.drop('Price', axis=1)
y= df['Price']
```

```python
from sklearn.linear_model import LinearRegression
reg= LinearRegression()
reg.fit(df[['Weight']],y)
```

# compute the slope (M) and y-intercept (C)

```python
reg.coef_ #M
```

```python
reg.intercept_ #c
```

# prediction and residuals

```python
reg.predict([[6]])
```

```python
reg.predict(x) 
```

```python
df['predicted Price'] = reg.predict(x)
df.head()
```

```python
df['Residuals'] =abs(y - reg.predict(x))
df.head()
```

# calculating MSE & MAE

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
MSE = mean_squared_error(y, reg.predict(x))
MSE
```

```python
MAE = mean_absolute_error(y, reg.predict(x))
MAE
```

# visualization

```python
from matplotlib import pyplot as plt
plt.scatter(x, y, color='blue')
plt.xlabel('Weight')
plt.ylabel('Price')
plt.title('Weight-Price visualization')
plt.plot(x, reg.predict(x))
```

```python
#scaling in ml
import pandas as pd
import numpy as np
data={'Price':[110, 105, 115, 120, 110, 130, 150, 100, 103]}
df= pd.DataFrame(data)
df.head(9)
```

```python

#Normalization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['price_scaled'] = scaler.fit_transform(df[['Price']])
df.head(9)
```

```python
#Standard scaler
from sklearn.preprocessing import StandardScaler
std= StandardScaler()
df['price_scaled_std']= std.fit_transform(df[['Price']])
df.head(9)
```

```python
#Robust scaler
from sklearn.preprocessing import RobustScaler
df['price_scaled_robust']= RobustScaler().fit_transform(df[['Price']])
df.head(9)
```

```python
#Log Transform
from sklearn.preprocessing import FunctionTransformer
df['log_price']= FunctionTransformer(np.log1p).fit_transform(df[['Price']])
df.head(9)
```

```python
#Max absolute Scaler
from sklearn.preprocessing import MaxAbsScaler
scaler= MaxAbsScaler()
df['max_abs']= scaler.fit_transform(df[['Price']])
df.head(9)
```
